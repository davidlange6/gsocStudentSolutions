```python
def mass(*particles):
    energy = particles.map(lambda particle: particle.energy).sum
    px = particles.map(lambda particle: particle.px).sum
    py = particles.map(lambda particle: particle.py).sum
    pz = particles.map(lambda particle: particle.pz).sum
    return math.sqrt(energy**2 - px**2 - py**2 - pz**2)
```

<p>Mass of a Z Boson should be Approx 91Gev. ie- Pair of electron if comes from a Z boson it should be close to 91Gev and If pair of muons comes from Z boson it should be close to 91Gev. Now in each event if 2 Z Bosons are found then this necessary conditions could suggest that it was Higgs Boson.

# Return list of index who have min(4 leptons or 2 photons)
### 1) Return list of event index which have min( 4 leptons ) & contibute to 2 Z-Bosons
### 2) Return list of event index which have min( 2 photons ) & contibute to Higgs Boson

```python
leptons=(events
         .lazy
         .filter(lambda event: (event.electrons.size>=2 and event.muons.size>=2) 
                 or (event.electrons.size>=4)
                 or (event.muons.size>=4)))
    

photons=(events
         .lazy
         .filter(lambda event: (event.photons.size>=2)
                and not((event.electrons.size>=2 and event.muons.size>=2) 
                 or (event.electrons.size>=4)
                 or (event.muons.size>=4))))
```

```python
#### Get the index of events of leptons ###
leptons_index=leptons.collect

#### Get the index of events of photons ###
photons_index=photons.collect
```

```python
def mass(*particles):
    energy = particles.map(lambda particle: particle.energy).sum
    px = particles.map(lambda particle: particle.px).sum
    py = particles.map(lambda particle: particle.py).sum
    pz = particles.map(lambda particle: particle.pz).sum
    return math.sqrt(energy**2 - px**2 - py**2 - pz**2)
```

```python
Zmass=(leptons_index
       .lazy
       .map(lambda event: event.muons.pairs(mass) + event.electrons.pairs(mass))
       )
```

```python
Zmass=Zmass.collect
print "Events: ",Zmass.size
```

## Collect the index of Zmass calculated above with 2 Tolerance of 91Gev.

```python
possibleZmass_Index=[]
for i,sublist in enumerate(Zmass):
    flag=0
    for j,value in enumerate(sublist):
        if value-91>=-2 and value-91<=2 :
            flag+=1
    if flag>=2:
        possibleZmass_Index.append(i)    
```
